> Diabetes data analyzation

```python
db = pd.read_csv('diabetes.csv')
```

```python
db.sample()
```

```python
dx = db.iloc[:,[1,5,7]]
dy = db.iloc[:,-1]
```

```python
xt,XT,yt,YT = train_test_split(dx,dy,test_size = 0.3)
```

```python

DT = []
for i in range(20):
#     xt,XT,yt,YT = train_test_split(dx,dy,test_size = 0.3)
    model = DecisionTreeClassifier(min_samples_split=2)
    model =  model.fit(xt,yt)
    YTP = model.predict(XT)
    DT.append(metrics.accuracy_score(YT,YTP))
```

```python
RF = []
for i in range(20):
#     xt,XT,yt,YT = train_test_split(dx,dy,test_size = 0.3)
    model = RandomForestClassifier()
    model =  model.fit(xt,yt)
    YTP = model.predict(XT)
    RF.append(metrics.accuracy_score(YT,YTP))
```

```python
Bag = []
for i in range(20):
#     xt,XT,yt,YT = train_test_split(ddx,ddy,test_size = 0.3)
    mdl = bagging.BaggingClassifier(n_estimators = 500)
    mdl.max_samples=0.9
    mdl = mdl.fit(xt,yt)
    YTP = mdl.predict(XT)
    Bag.append(metrics.accuracy_score(YTP,YT))
```

```python
adab = []
for i in range(20):
    ada = AdaBoostClassifier(n_estimators=50,learning_rate=0.2)
    ada = ada.fit(xt,yt)
    YTP = ada.predict(XT)
#     print(metrics.accuracy_score(yt,ada.predict(xt)))
#     print(metrics.accuracy_score(YT,YTP))
    adab.append(metrics.accuracy_score(YT,YTP))
#     train.append(metrics.accuracy_score(yt,ada.predict(xt)))
```

```python
acdf = pd.DataFrame()
acdf['Acc'] = DT+RF+Bag+adab
acdf['C'] = ['DT']*20+['RF']*20+['Bag']*20+['Adaboost']*20
```

```python
plt.figure(figsize=(16,9))
sns.set_style('whitegrid')
sns.boxplot(acdf['C'],acdf['Acc'])
```

```python
%%time
DT = [];RF=[];Bag=[];adab=[]
for i in range(50):
    xt,XT,yt,YT = train_test_split(dx,dy,test_size = 0.3)
    model1 = DecisionTreeClassifier(min_samples_split=2)
    model1 =  model1.fit(xt,yt)
    YTP = model1.predict(XT)
    DT.append(metrics.accuracy_score(YT,YTP))
    model2 = RandomForestClassifier(1000)
    model2 =  model2.fit(xt,yt)
    YTP = model2.predict(XT)
    RF.append(metrics.accuracy_score(YT,YTP))
    mdl = bagging.BaggingClassifier(n_estimators = 1000)
#     mdl.max_samples=0.9
    mdl = mdl.fit(xt,yt)
    YTP = mdl.predict(XT)
    Bag.append(metrics.accuracy_score(YTP,YT))
    ada = AdaBoostClassifier(n_estimators=1000,learning_rate=0.2)
    ada = ada.fit(xt,yt)
    YTP = ada.predict(XT)
    adab.append(metrics.accuracy_score(YT,YTP))
```

```python
acdf2 = pd.DataFrame()
acdf2['Acc'] = DT+RF+Bag+adab
acdf2['C'] = ['DT']*50+['RF']*50+['Bag']*50+['Adaboost']*50
```

```python
plt.figure(figsize=(16,9))
sns.set_style('whitegrid')
sns.boxplot(acdf2['C'],acdf2['Acc'])
```
